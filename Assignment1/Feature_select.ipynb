```python
##IMPORTS
import sys
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.linear_model import LassoLars
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import mean_squared_error
```

```python
train_file='train.csv'
test_file='test.csv'
output_file='modelpredictions1.txt'
```

```python
train_df = pd.read_csv(train_file)
test_df = pd.read_csv(test_file)

target_column = train_df.columns[-1]

X_train = train_df.iloc[:, :-1]
y_train = train_df[target_column]
X_test = test_df
```

```python
low_cardinality_features = [col for col in X_train.columns if X_train[col].nunique() <= 5]

# One-Hot Encoding for low cardinality features
ohe = OneHotEncoder(drop='first', sparse_output=False)
ohe_train = pd.DataFrame(ohe.fit_transform(X_train[low_cardinality_features]), columns=ohe.get_feature_names_out(low_cardinality_features))
ohe_test = pd.DataFrame(ohe.transform(X_test[low_cardinality_features]), columns=ohe.get_feature_names_out(low_cardinality_features))
```

```python
errors = np.abs(y_train - np.median(y_train))
weights = np.where(errors > np.percentile(errors, 90), 0, 1.0)

X_train_filtered = X_train[weights == 1.0]
y_train_filtered = y_train[weights == 1.0]
```

```python
params={"alpha":np.arange(0.00001, 20, 500)}
Kf=KFold(n_splits=10, shuffle=False)
lasso=LassoLars()

lasso_cv=GridSearchCV(lasso, param_grid=params, cv=Kf)
lasso_cv.fit(X_train_filtered, y_train_filtered)
```

```python
print("Best Params {}".format(lasso_cv.best_params_))
```

```python
names=train_df.drop("Total Costs", axis=1).columns
print("Column Names: {}".format(names.values))
```

```python
import matplotlib.pyplot as plt
lasso1 = LassoLars(alpha=0.00001)
lasso1.fit(X_train_filtered, y_train_filtered)

# Using np.abs() to make coefficients positive.  
lasso1_coef = np.abs(lasso1.coef_)

# plotting the Column Names and Importance of Columns. 
plt.bar(names, lasso1_coef)
plt.xticks(rotation=90)
plt.grid()
plt.title("Feature Selection Based on Lasso")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.savefig('plot.png')
plt.show()
```

```python
feature_subset=np.array(names)[lasso1_coef>1000]
print("Selected Feature Columns: {}".format(feature_subset))

# Adding the target to the list of feaatures. 
feature_subset=np.append(feature_subset, "Total Costs")
print("Selected Columns: {}".format(feature_subset))
```

```python
df_new = train_df[feature_subset]
df_test_new = test_df[feature_subset[:-1]]
```

```python
len(feature_subset)
```

```python
X_train_selected = df_new.iloc[:, :-1]
X_test_selected = df_test_new
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_selected)
X_test_scaled = scaler.transform(X_test_selected)
```

```python
poly = PolynomialFeatures(degree=4, interaction_only=False, include_bias=False)
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)
print(X_train_poly.shape)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train_poly, y_train, sample_weight=weights)
```

```python
y_pred = linear_reg.predict(X_test_poly)
```

```python
np.savetxt(output_file, y_pred, fmt='%.6f')
```
